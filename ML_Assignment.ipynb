#2.What is correlation?

Correlation measures the relationship between two variables ‚Äî how one changes when the other changes.

Negative correlation means that when one variable increases, the other decreases.

#6. What are continuous and categorical variables?

Continuous variables: Numeric values that can take any value in a range (e.g., height, temperature).

Categorical variables: Represent categories or labels (e.g., gender, color).

#17. What is causation? Explain difference between correlation and causation with an example.

Causation means one variable actually causes a change in another.

Correlation: Ice cream sales and drowning cases increase together.

Causation: Hot weather causes both ‚Äî not ice cream!

#23. What is feature scaling? How does it help in Machine Learning?

Feature scaling standardizes the range of data features so that no feature dominates others.
Helps algorithms like SVM, KNN, Gradient Descent converge faster.
